# Introduction to Jupyter notebooks

The Jupyter notebooks offer a web-based environment for creating and
sharing computational documents. They enable a simple, streamlined,
document-centric workflow, and have thus seen wide adoption in teaching
and data science.

Key features of Jupyter notebooks include the capability to include and
execute Python (and other) code, together with Markdown-formatted text.
Markdown enables subsequently typesetting mathematical expressions,
using the LaTeX syntax.

There is a vast amount of information, documentation and tutorials on
Jupyter notebooks and Markdown available on the internet, and a good
starting point is the documentation at [Project Jupyter
Documentation](https://docs.jupyter.org/). In particular, the tutorial:
[Introduction to the JupyterLab and Jupyter
Notebooks](https://jupyter.org/try-jupyter/notebooks/?path=notebooks/Intro.ipynb)

## Markdown

[Markdown](https://daringfireball.net/projects/markdown/) was created as
a very simple text-to-html conversion tool, but has ever since found
widespread adoption as a general, yet lightweight, markup language for
creating formatted text.

Markdown is trivial to learn, and a basic introduction can be found in
the Jupyter notebook documentation: [Markdown
Cells](https://jupyter-notebook.readthedocs.io/en/latest/examples/Notebook/Working%20With%20Markdown%20Cells.html)

Typesetting mathematical expressions in Markdown is enabled using the
LaTeX syntax, e.g. using either “inline math” $a^2 + b^2 = c^2$ or for
larger equations “display mode” is better suited:

$$
       \dfrac{\partial u}{\partial x}
    +  \dfrac{\partial uu}{\partial x}
    =
      -\dfrac{1}{\rho} \dfrac{\partial p}{\partial x}
    +  \nu \dfrac{\partial^2 u}{\partial x^2}
$$

## Python code environment (i.e. cells)

Jupyter notebook are based on the idea to combine code together with
text describing and documenting it. The code is not only listed in the
document, but also executed to carry out interactive computations.
Jupyter notebook implementations natively support syntax highlighting,
ensuring readability and proper formatting of the code.

The following Python code, showing the simple plotting example from the
Python introduction lecture, can be modified and executed interactively
in a Jupyter notebook environment such as [CSC
Noppe](https://noppe.csc.fi/), [Google
Colab](https://colab.research.google.com/), [GitHub
Codespaces](https://github.com/codespaces) or running a local JupyterLab
on your personal computer:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
x = np.linspace(0, 2*np.pi, 101)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure
fig, ax = plt.subplots(layout="constrained")

# Plot the data
ax.plot(x, y1, label='sin(x)')
ax.plot(x, y2, label='cos(x)')

# Set labels and add a legend
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend(loc='best')

fig.show()
```

## Plotting

Figures are automatically included in Jupyter notebooks when they are
created in Python code. Those figures are updated automatically with
every new execution of the code cell. Alternatively, static images,
e.g. in PNG or JPEG format, can also be included in text cells. The
latter allows, e.g., to include hand written solutions to some of the
homework problems.
